# Data Exploration

In this section, we will explore the dataset to understand its structure, the distribution of images across different splits (train, test, and validation), and the disease categories.

## Dataset Overview
The dataset contains images of guava leaves categorized into three classes based on their condition:
1. **Healthy Guava**
2. **Fruit Fly**
3. **Anthracnose**

The data is split into three subsets:
- **Training Set**: Used for model training.
- **Validation Set**: Used to tune model hyperparameters.
- **Test Set**: Used to evaluate the model's performance.

## Data Distribution
Below is the distribution of images across different subsets and categories:

| Data Split | Category       | Image Count |
|------------|----------------|-------------|
| Val        | Healthy Guava  | 185         |
| Val        | Fruit Fly      | 262         |
| Val        | Anthracnose    | 308         |
| Test       | Healthy Guava  | 94          |
| Test       | Fruit Fly      | 132         |
| Test       | Anthracnose    | 156         |
| Train      | Healthy Guava  | 649         |
| Train      | Fruit Fly      | 918         |
| Train      | Anthracnose    | 1080        |

### Insights:
1. The dataset is well-organized into training, validation, and test splits.
2. There are a total of **3 categories** (Healthy Guava, Fruit Fly, Anthracnose).
3. The training set has the largest number of images, which is crucial for effective model learning.
4. The validation and test sets have smaller, balanced subsets for unbiased performance evaluation.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the paths for each category
anthracnose_path = "/home/ankit/guava_diseases_detection/data/GuavaDiseaseDataset/train/Anthracnose/1_unsharp_clahe_augmented_2.png"
fruit_fly_path = "/home/ankit/guava_diseases_detection/data/GuavaDiseaseDataset/train/fruit_fly/1_unsharp_clahe_augmented_1.png"
healthy_guava_path = "/home/ankit/guava_diseases_detection/data/GuavaDiseaseDataset/train/healthy_guava/1_unsharp_clahe_augmented_2.png"

# Load the images
anthracnose_img = mpimg.imread(anthracnose_path)
fruit_fly_img = mpimg.imread(fruit_fly_path)
healthy_guava_img = mpimg.imread(healthy_guava_path)

```

```python
# Set up the figure
plt.figure(figsize=(12, 6))

# Display Anthracnose image
plt.subplot(1, 3, 1)
plt.imshow(anthracnose_img)
plt.title("Anthracnose")
plt.axis('off')

# Display Fruit Fly image
plt.subplot(1, 3, 2)
plt.imshow(fruit_fly_img)
plt.title("Fruit Fly")
plt.axis('off')

# Display Healthy Guava image
plt.subplot(1, 3, 3)
plt.imshow(healthy_guava_img)
plt.title("Healthy Guava")
plt.axis('off')

# Adjust layout and show
plt.tight_layout()
plt.show()
```
